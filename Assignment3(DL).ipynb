```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("hussainnasirkhan/multiple-linear-regression-dataset")

print("Dataset path:", path)
print("Files inside folder:", os.listdir(path))
```

```python
import pandas as pd
import numpy as np
```

```python
print(os.listdir(path))
```

```python
df = pd.read_csv(os.path.join(path, "multiple_linear_regression_dataset.csv"))

print(df.head())
print(df.columns)
print(df.shape)
```

```python
X = df[["experience", "age"]].values
y = df["income"].values
```

```python
import numpy as np

n_features = X.shape[1]

w = np.zeros(n_features)
b = 0.0
```

```python
def predict(X, w, b):
    y_hat = X.dot(w) + b
    return y_hat
```

```python

def mean_squared_error(y, y_hat):
    loss = ((y_hat - y) ** 2).mean()
    return loss
```

```python
def compute_gradients(X, y, y_hat):
    N = len(y)
    dw = (2 / N) * X.T.dot(y_hat - y)
    db = (2 / N) * (y_hat - y).sum()
    return dw, db
```

```python
def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
lr = 0.0001
epochs = 1000

for epoch in range(epochs):
    y_hat = predict(X, w, b)
    loss = mean_squared_error(y, y_hat)
    dw, db = compute_gradients(X, y, y_hat)
    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print("Epoch:", epoch, "Loss:", loss)
```

```python
print("Final Weights:", w)
print("Final Bias:", b)

new_candidate = np.array([4.5, 68])
predicted_salary = new_candidate.dot(w) + b

print("Predicted Salary:", predicted_salary)
```

Concept Overview

Multiple Linear Regression using a linear perceptron predicts continuous values like salary

The model computes a weighted sum of input features and adds a bias:
There is no activation function because regression requires real-valued outputs.
It is essentially a perceptron used for numeric prediction instead of classification.

Learning Mechanism

The model learns by minimizing Mean Squared Error (MSE).
MSE measures the average squared difference between predicted and actual values.
Using gradient descent, weights and bias are updated gradually to reduce error.
As training progresses, loss decreases and predictions become more accurate.

Analysis and Limitations

The model provides smooth and continuous predictions.
However, it assumes a linear relationship between inputs and output.
If the relationship is nonlinear, performance may reduce.
