```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("/content/multiple_linear_regression_dataset.csv")

print("First 5 rows:\n", data.head())
print("\nColumns:", data.columns.tolist())
print("\nDataset Shape:", data.shape)
```

```python
X = data[["age", "experience"]].values
y = data["income"].values

print("\nShape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
n_features = X.shape[1]
w = np.zeros(n_features)
b = 0.0
```

```python
def predict(X, w, b):
    return np.dot(X, w) + b
```

```python
def mean_squared_error(y, y_hat):
    return np.mean((y_hat - y) ** 2)
```

```python
def compute_gradients(X, y, y_hat):
    N = len(y)
    error = y_hat - y

    dw = (2 / N) * np.dot(X.T, error)
    db = (2 / N) * np.sum(error)

    return dw, db
```

```python
def update_parameters(w, b, dw, db, lr):
    w -= lr * dw
    b -= lr * db
    return w, b
```

```python
lr = 0.01
epochs = 1000

print("\nTraining Started...\n")

for epoch in range(epochs):

    # Forward pass
    y_hat = predict(X, w, b)

    # Compute loss
    loss = mean_squared_error(y, y_hat)

    # Compute gradients
    dw, db = compute_gradients(X, y, y_hat)

    # Update parameters
    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print(f"Epoch {epoch:4d} | Loss: {loss:.4f}")

print("\nTraining Completed!")
```

```python
print("\nFinal Weights:", w)
print("Final Bias:", b)
```

```python
new_candidate = np.array([[4.5, 68]])   # 4.5 years experience, score 68
new_candidate_scaled = scaler.transform(new_candidate)

predicted_salary = predict(new_candidate_scaled, w, b)

print("\nPredicted Salary for (4.5 yrs, 68 score):", predicted_salary[0])
```
